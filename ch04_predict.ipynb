# 4章 予測関数

- 「レイヤー関数」を積み重ねて予測する
- Yp = W * X + B が `nn.Linear`
- PyTorch プログラミングにおける「レイヤー関数」
  - 脳の神経細胞の接続みたいな絵の丸いノードは中間テンソル。「レイヤー関数」はその間に挟まるもの
  - image.png

```python
import torch
import torch.nn as nn
```

```python
l1 = nn.Linear(784, 128)
# 「128」が脳神経細胞の絵の「隠れ層」の個数
l2 = nn.Linear(128, 10)
relu = nn.ReLU(inplace=True)
```

```python
inputs = torch.randn(100, 784)
m1 = l1(inputs)
m2 = relu(m1)
outputs = l2(m2)
print(outputs.shape)
```

```python
# 全体を合成関数として定義
net2 = nn.Sequential(l1, relu, l2)
outputs2 = net2(inputs)
print(outputs2.shape)
```

- 予測関数、損失関数、最適化関数の３つを意識
